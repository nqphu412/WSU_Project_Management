## PMBOK
- 5 project management process groups:

  - Need check-point: monitor, control, track, review, regulate, modify, adapt to avoid failure
- 10 knowledge areas:
    - **Integration**: processes and activities to identify, define, combine, unify, and coordinate the various processes and project management activities
    - **Scope**: processes to ensure that the project includes all the work required, and only the work required, to complete the project successfully
    - **Schedule**: processes to manage timely completion of the project
    - **Cost**: processes involved in planning, estimating, budgeting, financing, funding, managing, and controlling costs so that the project can be completed within the approved budget
    - **Quality**: processes and activities of the performing organization that determine quality policies, objectives, and responsibilities so that the project will satisfy the needs for which it was undertaken
    - **Resource**: processes that organize, manage, and lead the project team
    - **Communications**: processes to ensure timely and appropriate planning, collection, creation, distribution, storage, retrieval, management, control, monitoring, and ultimate disposition of project information --> We need this knowledge to ask the right question, choose best communication channel (email, written, meeting, online)
    - **Risk**: processes of conducting risk management planning, identification, analysis, response planning, and controlto increase the likelihood and impact of positive events and decrease the likelihood and impact of negative events in the project --> Back-up plan, know which scenario should be the one to act/change quickly (COVID)
    - **Procurement**: processes to purchase or acquire products, services, or results from outside the project team
    - **Stakeholder Management**: processes to identify the people, groups, or organizations, that could impact or be impacted by the project, analyze their
expectations and impact, and develop strategies for engaging them and managing conflicting interests -->

### Plan-driven Project Roles

- Executive roles are the one raising the decision FOR (BEFORE) the project
- Managerial roles are the one making the decision DURING the project (delay, need more fund, change, modify, when should be continue, which should be ignore to wish continue,...)
- Note: Subject Matter Expert (SME) only follow, not the one making final decision (They have expert in what they are doing, their area, but not the overall scenario of the entire project)

## AGILE
- **Individuals and interactions** over processes and tools
- **Working software** over comprehensive documentation
- **Customer collaboration** over contract negotiation
- **Responding to change** over following a plan

--> Talk to customers and hear what they need

### Agile Principles
- Customer satisfaction is given the highest priority.
- Unlike a plan-driven approach, changes in requirements areappreciated at any point in the project with a view to exploit competitive advantage for the customer. --> Entire system is flexible (timeline, budget)
- Present working software to the client frequently during the project execution phase. --> Checkpoint - if any changes need to be happen
- People representing business and development teams must work together constantly.
- Project team comprised of motivated people must be engaged in the project; support and trust must be extended to the project team.
- **Face-to-face communication** must be employed, which is considered effective and efficient. In other words, it is preferable to have a co-located project team.
- Project progress is the progress made in developing working software.
- Agile process is meant to create sustainable development, which demands sponsors, developers, and users work at the same pace for a long period.
- Uninterrupted attention to technical excellence and good design boosts agility.
- Simplicitythe art of maximizing the amount of work not done is essential.
- Self-managed teams help develop the best architectures, requirements, and designs.
- If not daily, the team routinely considers how to improve effectiveness and amends its behavior accordingly.


Keyword: Iteration

### Agile Scrum Role
- Product owner:  person responsible for the business value of the project and for deciding what work to do and in what order, as documented in the product backlog -> Make sure to bring the best finished result for the end users (Set direction for the project)
- Scrum Master (kinda project manager): person who ensures that the team is productive, facilitates the daily Scrum, enables close cooperation across all roles and functions, and removes barriers that prevent the team from being effective
- Scrum team or development team: cross-functional team of five to nine people (developer, tester, designer) who organize themselves and the work to produce the desired results for each sprint (milestone), which normally lasts two to four weeks

### Agile Artifacts
- Product backlog: list of features prioritized by business value (that list comes from the stakeholders, called user-stories, wishlist)
- Sprint backlog: highest-priority items from the product backlog to be completed within a sprint
- Burndown chart: shows the cumulative work remaining in a sprint on a day-by-day basis (usually toward 0 - so see the trend, do some linear regression to estimate the finish time)

### Agile Ceremonies
- Sprint planning session: meeting with the team to select a set of work from the product backlog to deliver during a sprint
- Daily Scrum: short meeting for the development team to share progress and challenges and plan work for the day
- Sprint review: meeting in which the team demonstrates to the product owner what it has completed during the sprint
- Sprint retrospective: meeting in which the team looks for ways to improve the product and the process based on a review of the actual performance of the development team

## PRINCE2: r Projects IN Controlled Environments
- structured project management methodology widely used in government and private sectors.
- focuses on best practices, roles, and processes to ensure successful project execution

### Principles (Kinda similar to PMBok)
- Continued Business Justification  A project must have a valid reason throughout its lifecycle.
- Learn from Experience  Past lessons are used to improve future projects.
- Defined Roles and Responsibilities  Every team member knows their duties.
- Manage by Stages  Projects are divided into manageable phases.
- Manage by Exception  Only escalate issues when needed.
- Focus on Products  Deliverables are prioritized over activities.
- Tailor to Suit the Project Environment  Adapt PRINCE2 to fit different industries or organizations.

### Them
- Business Case - Why is the project needed?
- Organization - Who is responsible for what?
- Quality - What standards must be met?
- Plans - How will tasks be carried out?
- Risk - What could go wrong?
- Change - How do we manage changes?
- Progress - How do we track success?

### Process
- Starting Up a Project (SU)  Initial planning and feasibility checks
- Directing a Project (DP)  Senior management oversight.
- Initiating a Project (IP)  Defining the detailed project scope.
- Controlling a Stage (CS)  Managing and monitoring individual phases.
- Managing Product Delivery (MP)  Ensuring deliverables meet quality standards.
- Managing a Stage Boundary (SB)  Assessing progress before moving to the next stage.
- Closing a Project (CP)  Final review, lessons learned, and formal closure.

PRINCE2 is primarily a predictive methodology, but it also has adaptive elements.

#### Predictive (Waterfall) Methodology
PRINCE2 follows a structured, process-driven approach with well-defined stages, roles, and documentation.It is best suited for projects with clear requirements and fixed scope, where detailed upfront planning is essential.
- Detailed project **plans created at the start**.
- Phased, sequential process with formal stage gates.
- Heavy focus on governance, risk management, and documentation.

#### Adaptive (Agile) Methodology
PRINCE2 can be tailored to work in iterative and flexible environments, especially with PRINCE2 Agile, which integrates Agile concepts like Scrum, Kanban, and Lean.
- The principle of "Tailor to Suit the Environment" allows PRINCE2 to be customized for Agile projects.
- **Manage by Exception approach** allows teams to work with autonomy while senior management intervenes only when necessary.
- **Focus on deliverables** (products) rather than strict adherence to processes.

##### PRINCE2 Agile Roles
- Agile Project Manager: Manages the project using Agile practices within the PRINCE2 framework.
- Team Members: Cross-functional teams that collaborate and adapt to changing requirements.
- Product Owner: Ensures the product meets customer needs and communicates project progress.
- Agile Coach: Provides guidance on Agile techniques and facilitates Agile ceremonies.

Benefits and Challenges:
- Benefits: Flexibility, better risk management, faster delivery, enhanced collaboration, customer-focused outcomes.
- Challenges: Potential for confusion between Agile and traditional PRINCE2 roles, the need for Agile expertise, and resistance to change in organizations.
