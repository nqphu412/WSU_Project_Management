### System:
System is a collection of interrelated components that work together to achieve a goal that is operated in a broad organisational environment.
- Do want to do anything outside of the system scope
- Project managers need to use **systems thinking**: Taking a holistic view of carrying out projects within the context of the organization

#### Organisational structure
It is more than just a hierarchical structure (Most common):
- Functional (hierarchical): CEO, functional managers for each department (HR, Account, IT, Engineering, ...)
    - Pros: share same expertise people in the same area (department), unity of command - 1 instruction made, faster communication due to the share of similar mindset, functional managers get overall ideas what they're doing
    - Cons: Self-esteem in themselves so might lead to slow decision-making, more time to contact to people in different area (through managerS)

- Project: CEO with program managers under (EY, PwC, IBM). Work and employ staff for current project. When finish, they move to another one or they're on bench (do nothing, just waiting for new project coming)
    - Pros: unity of command, traditional department barriers are reduced - variety of people with different skillset stay in the same area, more perspective for more views, common goal (of project), ... 
    - Cons: power struggle (high self-esteem) - conflict, lack of deeper expertise/understanding (guildance), cost of assignning member to part-time project (cost = time-consuming to get used to/collaboration), less communication productivity, poor reflection because fail to communicate lessons learned
 
- Matrix: middle ground between functional and project structures
    - Pros: communication faster, high productivity due to two manager supervisor (more idea), Lessons learned shared effectively, Cooperation between departments
    - Cons: communication slower, 2 bosses - more conflicts, meetings, more challenging to control
